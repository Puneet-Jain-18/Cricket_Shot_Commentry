# Pre-Processing

## Seperating Features and Target

```python
raw_X = dataset.iloc[:, :-1]
y = dataset.iloc[:,-1]
```

## One-hot encoding

```python
y = pd.get_dummies(y,prefix='shotType')
```

## Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(raw_X)
```

```python
raw_X
```

```python
X
```

# Train & Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Start building out Neural Net Model

## Importing Libraries: using Sequential model of keras

```python
from keras import Sequential
from keras.layers import Dense
```

```python
input_dim = X_train.shape[1]
```

```python
input_dim
```

```python
y_test.shape, y_train.shape
```

## Building Classifier

```python
classifier = Sequential()
```

### First Hidden Layer

```python
classifier.add(Dense(input_dim//2, activation='relu', kernel_initializer='random_normal', input_dim=input_dim))
```

### Second Hidden Layer

```python
classifier.add(Dense(input_dim//4, activation='relu', kernel_initializer='random_normal', input_dim=input_dim//2))
```

### Output Layer

```python
classifier.add(Dense(y.shape[1], activation='softmax', kernel_initializer='random_normal', input_dim=input_dim//4))
```

## Compiling the Neural Network

```python
classifier.compile(optimizer ='adam',loss='binary_crossentropy', metrics =['accuracy'])
```

## Fitting the data to the training dataset

```python
classifier.fit(X_train,y_train, batch_size=10, epochs=100)
```

# Evaluating Model

```python
eval_model = classifier.evaluate(X_train, y_train)
eval_model
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test.values.argmax(axis=1), y_pred.argmax(axis=1))
```
