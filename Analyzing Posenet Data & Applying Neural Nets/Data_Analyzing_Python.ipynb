# Pre-Processing

```python
raw_X = dataset.iloc[:, :-1]
y = dataset.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(raw_X)
```

```python
raw_X
```

```python
X
```

# Train & Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Start building out Neural Net Model

## Importing Libraries: using Sequential model of keras

```python
from keras import Sequential
from keras.layers import Dense
```

```python
input_dim = X_train.shape[1]
```

```python
input_dim
```

## Building Classifier

```python
classifier = Sequential()
```

### First Hidden Layer

```python
classifier.add(Dense(input_dim//2, activation='relu', kernel_initializer='random_normal', input_dim=input_dim))
```

### Second Hidden Layer

```python
classifier.add(Dense(input_dim//8, activation='relu', kernel_initializer='random_normal'))
```

### Output Layer

```python
classifier.add(Dense(6, activation='sigmoid', kernel_initializer='random_normal'))
```

## Compiling the Neural Network

```python
classifier.compile(optimizer ='adam',loss='binary_crossentropy', metrics =['accuracy'])
```

## Fitting the data to the training dataset

```python
classifier.fit(X_train,y_train, batch_size=10, epochs=100)
```
